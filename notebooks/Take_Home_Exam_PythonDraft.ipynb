Introduction:

Main research question: Is a higher consumption of alcohol associated with higher adult mortality rate within the Benelux region?

Split up into two parts:

*   What association is there within countries between Alcohol Consumption and Adult Mortality?
*   How do countries within the Benelux compare to each other?

To answer this research question, the following dataset will be used: "https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv"



The following Variables will be used:


*   Adult Mortality: Adult Mortality Rates of both sexes (probability of dying between 15 and 60 years per 1000 population)
*   Alcohol: recorded per capita (15+) consumption in litres of pure alcohol

The following countries are within the Benelux region and will be used:

*   Belgium
*   The Netherlands

*   Luxembourg

The data is first imported and then split off into the various countries, cleaned up and saved.


```python
##research question: Is a higher BMI associated with a lower life expectancy within the European Union?
##split up into two parts: What association is there within countries between BMI and life expectancy? How do countries within the European Union compare to each other?
import pandas as pd
HF = pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv")
HF.shape
HF.head()
```
